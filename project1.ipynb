```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn import metrics
```

```python
dataset = pd.read_csv('/content/heart_disease.csv')
```

```python
dataset.head(40)
```

```python
dataset['DEATH_EVENT'].value_counts()
```

```python
dataset.shape
```

```python
X = dataset.drop(['DEATH_EVENT'], axis=1)
Y = dataset['DEATH_EVENT']
```

```python
scale = StandardScaler()
scaled_X = scale.fit_transform(X)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(scaled_X,Y,random_state=42, test_size =0.2)
```

```python
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
#training on LogisticRegression
prediction_on_training_data = model.predict(X_train)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
print('Accuracy on training data : ', accuracy_on_training_data)
#
```

```python
#testing on Logistic Regression
predictions_on_testing_data = model.predict(X_test)
accuracy_on_testing_data = metrics.accuracy_score(Y_test,predictions_on_testing_data)
print('Accuracy on testing data : ', accuracy_on_testing_data)
```

```python
model1 = RandomForestClassifier()
```

```python
model1.fit(X_train,Y_train)
```

```python
# Training on RandomForest
prediction_on_training_data= model1.predict(X_train)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
print('Accuracy on training data : ', accuracy_on_training_data)
```

```python
# Evaluate RandomForest on test data
predictions_on_testing_data_svc = model1.predict(X_test)
accuracy_on_testing_data_svc = metrics.accuracy_score(Y_test, predictions_on_testing_data_svc)
print('Accuracy of SVC on testing data:', accuracy_on_testing_data_svc)
```

```python
model2 =SVC(probability=True)
```

```python
model2.fit(X_train,Y_train)
```

```python
# Training on SVC
prediction_on_training_data= model2.predict(X_train)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
print('Accuracy on training data : ', accuracy_on_training_data)
```

```python
#testing on SVC
predictions_on_testing_data_lr = model2.predict(X_test)
accuracy_on_testing_data_lr = metrics.accuracy_score(Y_test, predictions_on_testing_data_lr)
print('Accuracy of Logistic Regression on testing data:', accuracy_on_testing_data_lr)
```
